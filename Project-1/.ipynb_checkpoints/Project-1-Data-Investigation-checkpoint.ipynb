### Part 2: Real World Representation
For our 3 columns we'll look at `voc`, `analog_input`, and `pm2.5_alt`.

`voc`: VOC stands for volatile organic compounds which are usually human-made chemicals that are emitted as gases. This number represents the concentrations of VOCs in the air.

`analog_input`: Analog input is the voltage detected by the AirPurple sensors. However, only the newer models can measure this and the older ones return 0.

`pm2.5_alt`: pm2.5 referes to particulate matter or particle pollution at the scale of 2.5um and smaller. ALT means the pm2.5 value was calculated with the alternate formula as opposed to the ATM formula.
