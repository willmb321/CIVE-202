```python
### Install packages
import pandas as pd

### Load and setup data
air_data = pd.read_csv('AirQuality_Daily_StudentVersion.csv')
air_data = pd.DataFrame(air_data)
air_data.head(1)
```

```python
# get min values for each sensor
sort_minimum = air_data.groupby(['sensor.name']).agg(min=('pm2.5_atm', 'min'))
sort_minimum
```

```python
# Number of entries
temp = air_data[(air_data['sensor.name']=="Broken Bow")].groupby(['sensor.name']).size()
temp
```

```python
# get mode for each sensor
sort_mode = air_data.groupby(['sensor.name'])['pm2.5_atm'].agg(pd.Series.mode).to_frame()
sort_mode
```
